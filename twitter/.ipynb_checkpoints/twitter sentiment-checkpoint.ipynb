```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
pos_tag([w])
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops, string.punctuation
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document), category) for document, category in documents]
```

```python
training_documents_orig = documents
testing_documents_orig= test_documents

training_documents = documents[0:7000]
testing_documents = documents[7000:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
all_words_orig = []
for doc in training_documents_orig:
    all_words_orig += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
import nltk
freq = nltk.FreqDist(all_words_orig)
common = freq.most_common(3000)
features_orig = [i[0] for i in common]
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```
